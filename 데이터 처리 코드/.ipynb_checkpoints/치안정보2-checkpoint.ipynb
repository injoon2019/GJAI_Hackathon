```python
def get_centorid_data():
    # 구별 위치 정보 든 geo json 불러옴 --------------------------------------
    with open("korea_seperate_gu.json", encoding="utf-8-sig") as file:
        json_geo_data = json.load(file)

    list_dic_geo = []
    for dic_data in json_geo_data["features"]:
        # 전국 데이터에서 광주 구별 데이터만 ID로 가져옴.
        if dic_data["properties"]["SIG_CD"] in ["29170", "29200", "29140", "29155", "29110"]:
            list_dic_geo.append(dic_data)

    dic_all_centroid = {}
    # 각 구별로 계산 --------------------------------------
    for dic_distrcit_data in list_dic_geo:
        name = dic_distrcit_data["properties"]["SIG_KOR_NM"] # 남구 동구...
        # if name != "남구":
        #     continue

        dic_result = centroid.all_centroied_poses_of_district(dic_distrcit_data, div_num=15)
        dic_all_centroid[name] = dic_result

    return dic_all_centroid
```

```python
# DATA lights -------------------------------------
df_lights = pd.read_csv(r"C:\Users\injoo\광주 인공지능 사관학교\웜업 프로젝트\해커톤\데이터 처리 코드\cleaned_lights2.csv")
df_light = df_lights[["보안등위치명", "위도", "경도"]]

# DATA police -------------------------------------
df_police = pd.read_csv(r"C:\Users\injoo\광주 인공지능 사관학교\웜업 프로젝트\해커톤\데이터 처리 코드\police_data.csv")
```

```python
df_lights = df_lights[df_lights['위도'].notna()]
df_lights = df_lights[df_lights['경도'].notna()]
df_lights.to_csv("cleaned_lights2.csv", index=False)
df_lights.isnull().sum()
```

```python
# MARKER - republic library -------------------------------------
start_x = 35.139737
start_y = 126.913963
    
map = folium.Map(location=[start_x, start_y], zoom_start=10)
map_cluster_lights = folium.plugins.MarkerCluster()
map_cluster_police = folium.plugins.MarkerCluster()

# MARKER - light -------------------------------------
for idx in range(len(df_light)):
    series = df_light.loc[idx,:]
    marker = folium.Marker([series["위도"], series["경도"]],
                               popup=series["보안등위치명"],icon=folium.Icon(icon='book', color='red'))
    map_cluster_lights.add_child(marker)
    map.add_child(map_cluster_lights)


# MARKER - police -------------------------------------
for idx in range(len(df_police)):
    series = df_police.loc[idx,:]
    marker = folium.Marker([series["위도"], series["경도"]],
                               popup=series["명칭"],icon=folium.Icon(icon='book', color='green'))
    map_cluster_police.add_child(marker)
    map.add_child(map_cluster_police)

# CREATE centroid
dic_all_centroid = get_centorid_data()


```
